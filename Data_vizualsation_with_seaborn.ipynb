Seaborn is based on matplotlib that provides the high level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex vizualization with just a few lines of code , In this lesson we'll cover the basics of seaborn ,including creating various types of plots and custamizing them...

```python
import seaborn as sns

## Basic plotting by seaborn

tip=sns.load_dataset("tips")
```

```python
tip
```

```python
##creating a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y='tip',data=tip)
plt.title("Scatter Plot of total bill and tip")
```

```python
##Line plot

sns.lineplot(x='size',y="total_bill",data=tip)
plt.title("Line Plot between Size and Tip")
```

```python
## categorical plot(uses the categorical variable)

## bar plot

sns.barplot(x="day",y="total_bill",data=tip)
plt.title("Bar Plot of Day and Total Bill")

```

**BOX PLOT**

```python
##BOX PLOT
sns.boxplot(x="day",y="total_bill",data=tip)
```

**VIOLIN PLOT**

```python
##Violent plot

sns.violinplot(x='day',y='total_bill',data=tip)
```

**HISTOGRAM**


```python
sns.histplot(tip['total_bill'],bins=10,kde=True)
```

```python
sns.histplot(tip['total_bill'],bins=10,kde=False)
```

**KDE PLOT**

```python
sns.kdeplot(tip["total_bill"],fill=False)
```

```python
sns.kdeplot(tip["total_bill"],fill=True)
```

**PAIR PLOT**

```python
sns.pairplot(tip)
```

**HEAT MAP**

```python
## corr=correlation
corr=tip[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
