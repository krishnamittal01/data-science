```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

##create a line plot

plt.plot(x,y,color="green",linestyle=":",marker="*",linewidth=3,markersize=9)
plt.grid=True
plt.xlabel("X axis",color="blue")
plt.ylabel("Y axis",color="blue")
plt.title("Line Plot",color="blue")
plt.show()
```

**MULTIPLE PLOTS**

1. SUB PLOTS



```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(10,5))
#plt.subplot(row,column,grid(1st or 2nd))
plt.subplot(2,2,1)
plt.plot(x,y1,color="green")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color="red")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,x,color="blue")
plt.title("Plot 2")

plt.subplot(2,2,4)
plt.plot(y1,x,color="orange")
plt.title("Plot 4")

```

BAR PLOT

```python
categories=["A","B","C","D","E"]
values=[5,7,3,8,6]

##create a bar plot
plt.bar(categories,values,color="purple")
plt.xlabel("Categoris")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

**HISTOGRAM**

Histogram are used to represent the distribution of a dataset .They divide the data into bins and counts the number of data points in each bin.

```python
## sample data
data=[1,2,3,1,2,3,1,2,3,4,5]

## create histogram

plt.hist(data,bins=5,color="orange",edgecolor= "black")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

**SCATTER PLOT**

```python
## create a scatter plot
## sample data

x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color="blue",marker="*")
```

**PIE CHART**

```python
labels=["A","B","C","D"]
sizes=[30,20,40,10]
colors=["gold","yellowgreen","lightcoral","lightskyblue","red"]
explode=(0.2,0,0,0)


## create a pie chart

plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",shadow=True)
```
