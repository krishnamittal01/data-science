PANDAS

PANDAS IS A POWERFUL DATA MANUPULATION LIBRARY IN PYTHON , WIDELY USED FOR DATA ANALYSIS AND DATA CLEANING .IT PROVIDES TWO PRIMARY DATA STRUCTURES. SERIES AND DATA FRAMES A SERIES IS A ONBE DIMENSIONAL ARRAY LIKE OBJECT, WHILE THE DATA FRAME IS A 2D , SIZE-MUTABLE , AND POTENTIALLY HETEROGENOUS TABULAR DATA STRUCTURE WITH LABELED AXES (ROWS AND COLUMNS).


```python
import pandas as pd
```

```python
## Series
#pandas series is a 1-d array like object that can hold any type .It is simlilar to a column in a table
data=[1,2,3,4,5]
series=pd.Series(data)
print("Seirs \n:",series)
print(type(series))
```

```python
##create a series from dictionary element
data={'a':1,'b':2,"c":3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index) ### **Series exist in key value pairs**

```

Dataframe

```python
##create  Dataframe from a dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York','London']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
##Create a Data frame from the list of dictionaries

data=[
    {'Name':'Krish','Age':25,'City':'Bangalore'},
    {'Name':'John','Age':30,'City':'New York'},
    {'Name':'Jack','Age':45,'City':'London'}
]
df=pd.DataFrame(data)
print(df)
```
